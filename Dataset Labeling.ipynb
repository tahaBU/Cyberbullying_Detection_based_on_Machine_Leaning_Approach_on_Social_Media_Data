```python
import pandas as pd
url1= "/content/drive/MyDrive/Dataset/cyberbullyingalldata.csv"
url2= "/content/drive/MyDrive/Dataset/tweetData.csv"
```

```python
df1= pd.read_csv(url1)
df1
```

```python
df2= pd.read_csv(url2)
df2
```

```python
# merging two csv files
df = pd.concat(
    map(pd.read_csv, [url1, url2]), ignore_index=True)
df
```

```python
from textblob import TextBlob
from textblob.sentiments import NaiveBayesAnalyzer

df['label'] = ''
for i,x in df.Tweet.astype(str).iteritems():
    label = TextBlob(x)
    df['label'][i] = label.sentiment.polarity
    print("Index: ", i ,  "label" , label.sentiment.polarity)
```

```python
def polarity_to_label(x):
    if(x >= -1 and x < 0):
        return 1
    else:
        return 0
df.label = df.label.apply(polarity_to_label)
```
